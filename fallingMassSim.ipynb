```python
#FallingMass Simulation 

import numpy as np
import matplotlib.pyplot as plt

#constants

mass = 5

gravity = 9.8

initialVelocity = 0

dt = 0.1

#variables

t = 0.0

h = 100

v = initialVelocity

distanceMoved = 0

#defining initial values for kinetic and potential energy
Ekin = 0.5*mass*v**2
EPot = mass*gravity*h

#lists

time = []
kineticEng = []
potentialEng = []
```

```python
while h>0:

    time.append(t)
    kineticEng.append(Ekin)
    potentialEng.append(EPot)

    distanceMoved = v*dt + 0.5*gravity*(dt**2)
    v+= gravity*dt
    t+= dt
    h -= distanceMoved

    #calcuating kinetic and potential energy for each time step

    Ekin = 0.5*mass*v**2
    EPot = mass*gravity*h
    
```

```python
fig, ax = plt.subplots()
ax.scatter(time, kineticEng, label ='Kinetic', marker='x', color='r')
ax.scatter(time, potentialEng, label ='Potential', marker='x', color='g')

ax.set_xlabel('Time')
ax.set_ylabel('Energy in Joules')
ax.legend()
```

```python
from matplotlib.animation import FuncAnimation

#resetting the variables to initial conditions
h = 100
t = 0.0
v = initialVelocity
distanceMoved = 0

Ekin = 0.5*mass*v**2
EPot = mass*gravity*h

#lists

time = []

kineticEng = []
potentialEng =[]

fig,ax = plt.subplots()
ax.set_xlim(0,6)
ax.set_ylim(0,6000)
ax.set_xlabel('Time')
ax.set_ylabel('Energy in Joules')

kineticPlot, = ax.plot([],[],label='Kinetic', marker='x', color='r')
potentialPlot, = ax.plot([],[],label='Potential', marker='x', color='g')

def update(frame):
    global t, h, v, distanceMoved, Ekin, EPot

    if h>=0:
        time.append(t)
        kineticEng.append(Ekin)
        potentialEng.append(EPot)

        distanceMoved = v*dt+0.5*gravity*(dt**2)
        v+= gravity*dt
        t+=dt
        h-= distanceMoved

        #calcuating kinetic and potential energy for each time step

        Ekin = 0.5*mass*v**2
        EPot = mass*gravity*h

        kineticPlot.set_data(time, kineticEng)
        potentialPlot.set_data(time, potentialEng)
    return kineticPlot, potentialPlot

ani = FuncAnimation(fig, update, frames = 200, interval=100)

ani.save('anim.gif')

        



```

```python
#resetting the variables to initial conditions
h = 100
t = 0.0
v = initialVelocity
distanceMoved = 0
k = 0.05 # coefficient of resistance
#resetting the lists
time = []
height = []
velocityList =  []
while h>0:
    #start by calculating forc due to gravity F=mg
    forceGrav = mass*gravity
    #force due to air resistance -kv^2
    forceAir = -k*v**2

    #total force calculation
    totForce = forceGrav + forceAir

    # calculating acceleration for f=ma
    acc = totForce/mass

    #appending to lists
    time.append(t)
    height.append(h)
    velocityList.append(v)

    #Updating values
    distanceMoved = v*dt+0.5*gravity*(dt**2)
    v+= acc*dt
    t+=dt
    h-= distanceMoved

    
```

```python
fig,ax = plt.subplots()
ax.scatter(time,velocityList,label='Air Resistance', marker = 'x')
ax.set_xlabel('Time')
ax.set_ylabel('Velocity')

ax.legend()
```
